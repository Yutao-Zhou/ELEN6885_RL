```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
labels = ['jigsaw_unintended_bias', 'OxAISH-AL-LLM/wiki_toxic', 'wikitext']
base_means = [0.311744823774556, 0.360249796624528, 0.107460239481297]
base_std = [0.393169946390773, 0.412608916117655, 0.236073080005579]

detoxifyed_means = [0.0291997655555024, 0.0296960882436251, 0.0184947845661663]
detoxifyed_std = [0.12521923485779, 0.104926674737973, 0.0997283485053301]

base_means = np.round(base_means, 2)
base_std = np.round(base_std, 2)

detoxifyed_means = np.round(detoxifyed_means, 2)
detoxifyed_std = np.round(detoxifyed_std, 2)

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, base_means, width, yerr=base_std, label='Base Means')
rects2 = ax.bar(x + width/2, detoxifyed_means, width, yerr=detoxifyed_std, label='Detoxifyed Means')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_xlabel('Data Set', weight="bold")
ax.set_ylabel('Mean Toxicity', weight="bold")
ax.set_title('Fine-tuned on jigsaw_unintended_bias Data Set', weight="bold")
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()


def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(12, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)
autolabel(rects2)

fig.tight_layout()

plt.show()
```
